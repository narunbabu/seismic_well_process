```python
from petrel_well_file_readers import *
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import lasio
from flex_log import FlexXY,FlexLog
from seis_ampl_spectrum import *
folder=r'D:\SoftwareWebApps\Python\geophysics\inversion&spect_decomp\d11_data\\'
# folder=r"D:\Ameyem\python\inversion&spectraldecomp\d11_data\\"
# las_file=folder+'nec25_a1.las'
# las=lasio.read(las_file)

# import os
# for f in os.listdir(folder):
#     if 'd11' in f:
#         print('"'+f+'"',end=', ')
# # dir(os)
flist={"ext_wt":"d11_extracted_wlt_nec25a1.dat",  "ricker":"d11_ricker30hz.dat", "seis_act":"d11_seis_actual.dat", 
       "seis_dept":"d11_seis_depth.dat","seis_synth":"d11_seis_synth.dat", "synth_las":"d11_synth_tie_logs.las"}

from bruges.filters import ricker as ricker_
from scipy import signal

# def ricky(length, width,dt): # dt = sampling interval in sec
#     FS = 1/dt
#     return ricker_(0.256, 1/FS, width)
def ai2rc(ai_res):
    ai_res=np.append(ai_res,ai_res[-1])
    return (ai_res[1:]-ai_res[:-1])/(ai_res[1:]+ai_res[:-1])
def myfilter(dt,sig,fc=60,order=3): # dt = sampling interval in sec, sig=data, fc= cut-off frequency
    
    # T=(t[-1]-t[0])/1000
    # fs=n/T
    fs=1/dt
    # fc = 5  # Cut-off frequency of the filter
    w = fc / (fs / 2) # Normalize the frequency
#     order=3
    b, a = signal.butter(order, w, 'low')
    return signal.filtfilt(b, a, sig)
def freq_spec(dt,sig):
    amp_spec = np.fft.rfft(sig)
    
    f = np.fft.rfftfreq(len(sig), d=dt)
    P = 20 * np.log10(np.abs(amp_spec) )  #Power in Decibel Scale
    fig02 = plt.figure()
    ax = fig02.add_subplot(111)
    ax.plot(f, P,'m')
    ax.set_xlabel('frequency (Hz)', fontsize=14)
    ax.set_ylabel('power (dB)', fontsize=14)
    ax.set_xlim([0,250])
    ax.grid('on')
    return amp_spec,f
def logPlot(data4plots):
    npanels=len(data4plots)
    fig01 = plt.figure(figsize=(npanels*1.5,10))
    for i in range(npanels):
        ax = fig01.add_subplot(1,npanels,i+1)
        ax.plot(data4plots[i]['x'], data4plots[i]['y'], 'k')
        ax.fill_betweenx(data4plots[i]['y'], data4plots[i]['x'], 0,
                      data4plots[i]['x'] > 0, color='k', alpha=1.0)
        ax.set_xticks([])
        ax.set_title(data4plots[i]['name'])
        if len(data4plots[i]['yrange'])==2:
            ax.set_ylim(data4plots[i]['yrange'][0],data4plots[i]['yrange'][1])
```

```python
## Getting RC - Module
chkt_file=folder+"nec25_a1_chkt.dat"
chkt=read_chkt(chkt_file)
X=chkt['MD'].values
X.shape=len(X),1
y=chkt['TWT picked'].values
md2twt = LinearRegression().fit(X, y)

las=lasio.read(folder+flist["synth_las"])

flexlog = FlexLog(np.array([[0,0]]))
x,y=las['DEPT'],las['AI']
md,ai=flexlog.get_nonanXYs(x,y)
md.shape=len(md),1
twt=md2twt.predict(md)
t=-twt
aidt=np.mean(np.diff(t))*0.001
# dt=0.002
freq=30
rk30 = ricker_(0.26,aidt,freq)
trk30=range(len(rk30))*aidt
# plt.plot(rk30)
# print(rk30.shape)
# seis_synth['twt'],seis_act['twt'] #/np.max(seis_act['ampl']

# amp_spec,f=freq_spec(aidt,rk30)
# print('Frequency centered at {:3.2f} Hz.'.format(f[np.argmax(amp_spec)]))

rc=ai2rc(ai)
rc[np.isnan(rc)]=0
def calculateSynth(rk30, rc):
    calculated = signal.convolve(rk30, rc)
    return calculated[int(len(rk30)/2):1-int(len(rk30)/2)]

synth_rk_wt={}
synth_rk_wt['ampl']=calculateSynth(rk30, rc)
synth_rk_wt['twt']=range(len(synth_rk_wt['ampl']))*aidt+t[0]*0.001

# las=lasio.read(folder+flist["synth_las"])
# plt.plot(las['RESAMPLEDAI'])
# las.keys()

rk_wt=read_twocolfile(folder+flist["ricker"],nheadrows=37,colnames=['twt','ampl'])
ext_wt=read_twocolfile(folder+flist["ext_wt"],nheadrows=29,colnames=['twt','ampl'])
seis_act=read_twocolfile(folder+flist["seis_act"],nheadrows=2,colnames=['twt','ampl'])
seis_synth=read_twocolfile(folder+flist["seis_synth"],nheadrows=2,colnames=['twt','ampl'])
# ext_wlt

data4plots=[{'x':las['RESAMPLEDAI'][:len(t)],'y': t,'name':'Res.AI','yrange':[seis_synth['twt'][0],seis_synth['twt'][-1]]},
    {'x':ai,'y': t,'name':'AI','yrange':[seis_synth['twt'][0],seis_synth['twt'][-1]]},
#     {'x':rk_wt['ampl'], 'y':rk_wt['twt']*0.001,'name':'Ricker 30hz','yrange':[]}  ,
#     {'x':rk30, 'y':range(len(rk30))*aidt,'name':'Rick Calc','yrange':[]}  , 
#     {'x':ext_wt['ampl'], 'y':ext_wt['twt']*0.001,'name':'Extracted Wavelet','yrange':[]},
   {'x':rc, 'y':t,'name':'RC','yrange':[seis_synth['twt'][0],seis_synth['twt'][-1]]},
    {'x':seis_synth['ampl'], 'y':seis_synth['twt'],'name':'Petrel Synth','yrange':[seis_synth['twt'][0],seis_synth['twt'][-1]]},
    
    {'x':synth_rk_wt['ampl'], 'y':synth_rk_wt['twt'],'name':'Syn wbruges rk30','yrange':[synth_rk_wt['twt'][0],seis_synth['twt'][-1]*0.001]}  , 
     {'x':seis_act['ampl'], 'y':seis_act['twt'],'name':'Seismic','yrange':[seis_synth['twt'][0],seis_synth['twt'][-1]]},
    ]
 

petdt=(rk_wt['twt'][1]-rk_wt['twt'][0])/1000


new_tstart=np.ceil(t[0]) # time need to match with seismic time 
if new_tstart%2!=0:
    new_tstart+=1    
# sample_start=new_tstart-t[0]
# sample_start,new_tstart
################################### RC Resample module ####################
tnew=np.arange(new_tstart,t[-1]+0.001,petdt*1000)

## Wavelet for dt=0.002sec sampling rc
rcrdt=np.abs(tnew[0]-tnew[1])*0.001
freq=30
rk_res30 = ricker_(0.26,rcrdt,freq)

# plt.plot(t,las['AI']/max(las['AI']))
# plt.plot(t,rc)

## Direct RC resampling
rc_flex=FlexLog(np.column_stack((t,rc)))
rc_res=rc_flex.resampleXYs(tnew)
data4plots.append({'x':rc_res[:,1], 'y':tnew,'name':'Res. RC','yrange':[tnew[0],tnew[-1]]})
data4plots.append({'x':calculateSynth(rk_res30, rc_res[:,1]), 'y':tnew,'name':'Res. Synth','yrange':[tnew[0],tnew[-1]]})

## rc resampling after removing high frequencies
rc_low=myfilter(aidt,rc,fc=60,order=3)
rc_flex=FlexLog(np.column_stack((t,rc_low)))
rc_low_res=rc_flex.resampleXYs(tnew)
rc_low_res[0:4,1]=0
data4plots.append({'x':rc_low_res[:,1], 'y':tnew,'name':'Res.lfRC','yrange':[tnew[0],tnew[-1]]})
data4plots.append({'x':calculateSynth(rk_res30, rc_low_res[:,1]), 'y':tnew,'name':'RC lf. Synth','yrange':[tnew[0],tnew[-1]]})
```
